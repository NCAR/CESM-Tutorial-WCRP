# Basic Unix

## `cd` (Change Directory):

**Description**: 

The `cd` command is used to change the current working directory. By specifying the name of a directory as an argument, you can navigate to that directory.

**Example**:

The command:
```
cd /home/user/documents
```
changes the current directory to `/home/user/documents`.

The command:
```
cd ~
```
bring you to your `home` directory. 

____

## `ls` (List):

**Description**: 

The `ls` command is used to list the files and directories in the current directory or a specified directory. It provides information about the contents of the directory, such as filenames, permissions, sizes, and timestamps.

**Example**:

`ls` - Lists the files and directories in the current directory.
`ls -all /home/user/documents`  - Lists the contents of the "/home/user/documents" directory in a long format, showing detailed information about each file and directory.

___

## pwd (Print Working Directory):

**Description**: 

The pwd command displays the current working directory, which is the directory in which you are currently located within the file system. It helps you determine your current location in the directory tree.

**Example**:

pwd - Displays the current working directory, such as "/home/user."

___

## cp (Copy):

**Description**: 

Copies files or directories from a source location to a destination location. Useful for duplicating files or creating backups.

**Example**: 

cp [options] source destination


___

## rm (Remove):

**Description**: 

Deletes files or directories. Be cautious, as deleted files are typically not recoverable, and the operation is irreversible.

**Example**: 

rm [options] file(s)


___

## mv (Move):

**Description**: 

Description: Moves or renames files and directories. It can also be used to move files from one directory to another.

**Example**: 

mv [options] source destination

___

## more (File Content Viewer):


**Description**: 

Description: Allows you to view the content of a text file one page at a time. You can navigate through the file using keyboard commands.

**Example**: 

more [file]

___

## less (File Content Viewer):



**Description**: 
Description: Similar to more, it lets you view the content of text files one page at a time. less provides more advanced navigation features compared to more.

**Example**: 

Usage: less [file]
___

## grep (Global Regular Expression Print):

**Description**: 
Description: Searches for lines containing a specified pattern within one or more text files. Useful for text searching and pattern matching.

**Example**: 
grep [options] pattern [files]


These basic Unix commands are fundamental for navigating and managing files and directories in a Unix-like operating system. They are often used in combination to navigate the file system, list files, and confirm your current location within the directory structure.
